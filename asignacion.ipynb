```python
import numpy as np
```

```python
def outliers_out(array):

    an_array = array
    mean = np.mean(an_array)


    standard_deviation = np.std(an_array)


    distance_from_mean = abs(an_array - mean)

    
    not_outlier = distance_from_mean < standard_deviation
    no_outliers = an_array[not_outlier]

    for i,row in enumerate(an_array):
        for j,num in enumerate(row):
            if num not in no_outliers:
                an_array[i][j] = 5 * no_outliers.max()

    return an_array
```

```python
def metodo_radio(diccionario):

    array = outliers_out(np.array(list(diccionario.values())))

    max_row = np.max(array, 1)

    for i,key in enumerate(diccionario.keys()):
        diccionario[key] = list(np.round(array[i] / max_row[i],2))

    diccionario_array = np.array(list(diccionario.values()))

    radio = round(diccionario_array.min(),2)

    
    print(f'radio es {radio}')
    
    posiciones = [0] * len(diccionario_array)

    while radio <= diccionario_array.max() and not all(isinstance(x, str) for x in posiciones):

        for key in diccionario.keys():
            if key not in posiciones and radio in diccionario[key] and type(posiciones[diccionario[key].index(radio)]) == int:
                    posiciones[diccionario[key].index(radio)] = key

        radio = round(radio + 0.01,2)

    return posiciones
```
