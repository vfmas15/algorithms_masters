```python
import pandas as pd
import numpy as np
from itertools import permutations
```

```python
diction = { 
    "A" : [
    [1,2,3],
    [4,5,6],
    [7,8,9]
],

"B" : [
    [1,2,3],
    [4,5,6],
    [7,8,9]
],

"C" : [
    [1,2,3],
    [4,5,6],
    [7,8,9]
],

"D" : [
    [1,2,3],
    [4,5,6]
],
"E" : [
    [1,2,3],
    [4,5,6],
    [7,8,9]
],

"F" : [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]}
```

```python
def orden_multiplicacion(diccionario):

    combinaciones = []
    combinaciones_finales = []

    for key in diccionario.keys():
        combinaciones.append(key)

    # Lista de tuplas con todos los diferentes ordenes dde multiplicación
    combinaciones_iniciales =  list(permutations(combinaciones, len(combinaciones)-1))


    for i,item in enumerate(combinaciones_iniciales):
        combinaciones_iniciales[i] = list(combinaciones_iniciales[i])



    for i, array in enumerate(combinaciones_iniciales):
        
        for key in diccionario.keys():
            
            #Añade al inicio y al final la letra restante
            if key not in array:
                inicio = combinaciones_iniciales[i] + [key]
                fin = [key] + combinaciones_iniciales[i]

                combinaciones_finales.append(inicio)
                combinaciones_finales.append(fin)


    return combinaciones_finales


```

```python
def numero_operaciones(lista, diccionario):

    ordenes = [[] for val in range(len(diccionario))]

    multiplicaciones = [[] for val in range(len(diccionario)-1)]

    for array in lista:
        primero = len(diccionario[array[0]])
        
        for i, letter in enumerate(array):

            ordenes[i].append(letter)

            if i == 0:
                comparar = len(diccionario[letter][0])

            elif i != 0:

                multiplicaciones[i-1].append([primero, len(diccionario[letter]),len(diccionario[letter][0])])

                if comparar != len(diccionario[letter]):

                    multiplicaciones[i-1][-1] = float('NaN')

                else:
                    comparar = len(diccionario[letter][0])

    df = pd.DataFrame( np.array(ordenes).transpose() ) 

    for i,multiplicacion in enumerate(multiplicaciones):
        df[i + len(diccionario)] = multiplicaciones[i]

    # Elimina combinaciones repetidas
    df = df.drop_duplicates(subset=[num for num in range(len(ordenes))])

    # Elimina combinaciones no multiplicables
    df = df.dropna()

    return df
```
