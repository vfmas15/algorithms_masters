```python
from itertools import permutations
```

```python
def optima(peso_tiempo,tiempos,tiempo_limite):
    '''
    Te da una rama inicial considerada optima
    '''

    optima = []
    tiempo = 0

    while len(peso_tiempo) > 0:

        index = peso_tiempo.index(max(peso_tiempo))

        tiempo += tiempos[index]

        if tiempo > tiempo_limite:
            break
        else:

            optima.append(max(peso_tiempo))

            peso_tiempo.remove(peso_tiempo[index])

    return optima
```

```python
def sub_array_en_array(principal, subarray):
    '''
    Encuentra si hay algun subarray en un array
    '''
    len_b = len(subarray)

    for i in range(len(principal)):  
        if principal[i:i+len_b] == subarray:
             return True

    return False
```

```python
def array_en_lista(listas,array):

    for lista in listas:
        if sub_array_en_array(lista, array):
            return True
    
    return False
```

```python
def poda(arbol,rama_de_poda):
    '''
    Te deja el arbol bien chopeadito de ramas malas
    '''

    arbol_podado = []

    for rama in arbol:
        if not sub_array_en_array(rama,rama_de_poda):
            arbol_podado.append(rama)

    return arbol_podado
```

```python
def asignacion(pesos,tiempos,tiempo_limite):
    '''
    Te asigna la rama más optima y te da todos los combos
    '''

    peso_entre_tiempo = [ pesos[i]/tiempos[i] for i in range(0,len(pesos))]

    diccionario_tiempos = dict(zip(peso_entre_tiempo, tiempos))
    diccionario_pesos = dict(zip(peso_entre_tiempo, pesos))

    rama_optima = optima(peso_entre_tiempo,tiempos,tiempo_limite)
    rama_optima_tiempos = [diccionario_tiempos[val] for val in rama_optima]

    arbol = list(permutations(peso_entre_tiempo, len(peso_entre_tiempo)))

    for i, array in enumerate(arbol):
        arbol[i] = list(arbol[i])

    itertry = len(arbol)
    i = 0

    para_poda = []

    while i < itertry:

        tiempo_temp = 0
        suma_temp = 0
        poda_temp = []


        if not array_en_lista(para_poda,arbol[i]):

            for num in arbol[i]:

                tiempo_temp += diccionario_tiempos[num]
                suma_temp += num
                
                #print(f'tiempo_temp {tiempo_temp}')

                poda_temp.append(num)

                #print(f' el numero {num} con tiempo acumulado {tiempo_temp}')
                

                if tiempo_temp >= sum(rama_optima_tiempos):
                    #print(f'Poda!!! con temporal {poda_temp}')
                    para_poda.append(poda_temp)

                    arbol = poda(arbol,poda_temp)
                    break
                elif tiempo_temp <= sum(rama_optima_tiempos) and suma_temp >= sum(rama_optima):
                    rama_optima = poda_temp

        itertry = len(arbol)
        i += 1
    
    

    rama_optima = [diccionario_pesos[val] for val in rama_optima]

    for i, rama in enumerate(arbol):
        arbol[i] = [diccionario_pesos[val] for val in rama]

    for i, rama in enumerate(para_poda):
        para_poda[i] = [diccionario_pesos[val] for val in rama]

    return rama_optima
```
